```python
import gym
from DQN_Agent import DQNAgent
import numpy as np
import plotly.express as px
import time
```

```python
env = gym.make('CartPole-v1')
```

```python
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
state_size, action_size
```

```python
batch_size = 128
agent = DQNAgent(state_size, action_size)
```

```python
done = False
scores = []
episode = 0

while sum(scores[-10:]) < 4000:
    episode += 1
    
    state, _ = env.reset()
    state = np.reshape(state, [1, state_size])
    
    for step in range(1000):
        
        action = agent.act(state)
        next_state, reward, done, _, _ = env.step(action)
        
        next_state = np.reshape(next_state, [1, state_size])
        
        agent.remember(state, action, reward, next_state, done)
        
        state = next_state
        
        if done:
            agent.punish_last(8)
            print("episode: {},\t score: {},\t epsilon: {:.2}".format(episode, step, agent.epsilon))
            break
            
    scores.append(step)
    if len(agent.memory) > batch_size:
        agent.replay(batch_size)
        
    if episode % 100 == 0:
        agent.save(f"./models/{episode}.h5")
        
# agent.save("./models/best.h5")
```

```python
data = scores
fig = px.line(x=range(len(data)), y=data, labels={'x': 'Episode', 'y': 'Score'})
fig.show()
```
