```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

## Defining the KL Divergence Function

```python
def kl_divergence(p, q):
    return np.sum(np.where(p != 0, p * np.log(p / q), 0))
```

## Testing with different probability distributions

```python
eps = 1e-5
dist1 = np.random.randn(100000)+100
dist2 = np.random.randn(100000)+100
dist1 = dist1/np.linalg.norm(dist1)
dist2 = dist2/np.linalg.norm(dist2)

plt.figure(figsize=(8,4))
plt.hist(dist1,bins=100, label ="p", density = True,histtype = "step")
plt.hist(dist2,bins=100, label = "q", density = True,histtype = "step")
plt.plot([],[],label = "$D_{KL}(q||p)$ = "+f"{round(kl_divergence(dist2,dist1),5)}",ls = " ")
plt.plot([],[],label = "$D_{KL}(p||q)$ = "+f"{round(kl_divergence(dist1,dist2),5)}",ls = " ")
plt.legend()
plt.show()
```
